```python
import numpy as np
import matplotlib.pyplot as plt
import random

import math
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 10), dpi=80, facecolor='grey', edgecolor='pink')


x=[3]
y=[3]
end_x=5
end_y=5
total_ants=20


for i in range(5):
    
    ant_x=[]
    ant_y=[]
    distance=[]

    for i in range(20):
        ant_x.append(x[-1]+random.random()*.5)
        ant_y.append(y[-1]+random.random()*.5)
        plt.scatter(ant_x,ant_y)
        plt.xlim(3,5)
        plt.ylim(3,5)

    for i in range(20):
        distance.append(math.sqrt((math.pow((ant_x[i]-5),2))+(math.pow((ant_y[i]-5),2))))

    ant_x[distance.index(min(distance))]
    plt.scatter(ant_x[distance.index(min(distance))],ant_y[distance.index(min(distance))],color='red')   
    x.append(ant_x[distance.index(min(distance))])
    y.append(ant_y[distance.index(min(distance))])
    
plt.xlim(3,6)
plt.ylim(3,6)

plt.plot(x,y,color='blue')
```
